```python
import numpy as np
import matplotlib.pyplot as plt

def quadratic(x, a=1, b=0, c=0):
    return a*x**2 + b*x + c

x_values = np.linspace(-10, 10, 500)

y_values = quadratic(x_values)

plt.figure(figsize=(5, 5))
plt.plot(x_values, y_values)
plt.title('Plot of a Quadratic Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
